```python
import pandas as pd
import numpy as np
```

```python
years=[i for i in range(2005,2021,3)]
```

```python
years
```

```python
final = pd.DataFrame()
```

```python
for year in years:
    df=pd.read_csv('/home/nishant/study/researchwork/MSThesis-Time_Series_Forecasting-master/Data/original/Rainfall/Maharashtra/Solapur/India - Maharashtra - Solapur__Rainfall'+str(year)+'.csv')
    df=df[['Year','Month','Rainfall (mm)']]
    p=[]
    for i in range(0,len(df),3):
        x= df.loc[i]['Rainfall (mm)'] + df.loc[i+1]['Rainfall (mm)'] + df.loc[i+2]['Rainfall (mm)']
        p.extend([x]*3)
    df['Rainfall']=p
    dk=df[df.index.isin([i for i in range(0,108,3)])]
    dk=dk[['Year','Month','Rainfall']]
    dk.reset_index(inplace=True,drop=True)
    final=final.append(dk)
final.reset_index(inplace=True,drop=True)
final=final.replace(np.nan,0.0)
final.to_csv('Solapur_full.csv')
```
