```python
import pandas as pd
import numpy as np
```

```python
years=[i for i in range(2005,2021,3)]
```

```python
final = pd.DataFrame()
```

```python
df=pd.read_csv('India - Maharashtra - Ahmednagar__Rainfall2005.csv')
```

```python
df=df[df['Dekad']==1][['Year','Month','Rainfall (mm)']]
```

```python
for year in years:
    df=pd.read_csv('India - Maharashtra - Ahmednagar__Rainfall'+str(year)+'.csv')
    df=df[df['Dekad']==1][['Year','Month','Rainfall (mm)']]
    final=final.append(df)
final.reset_index(inplace=True,drop=True)
final=final.replace(np.nan,0.0)
```

```python
final.to_csv('ahmednagar_full.csv')
```
