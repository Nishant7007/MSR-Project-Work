```python
from datetime import timedelta
date_list=[]
initial=pd.to_datetime('2006-01-01')
final=pd.to_datetime('2019-12-31')
s=initial
e=s+timedelta(43)    
while(e<=final):
    x=df[1][s:e].max()/df[1][s:e].min()
    #print(s,e,x)
    if(x<1.4 and x>1.1):
        date_list.append([s,e,x])
        s=s+timedelta(44)
        e=e+timedelta(44)
        continue
    else:
        s=s+timedelta(7)
        e=e+timedelta(7)
```
